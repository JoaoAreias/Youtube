```python
import matplotlib.pyplot as plt
import solution
import numpy as np
```

```python
input1 = list(map(int, "1 2 3".split()))
input2 = list(map(int, "1 2 3 4".split()))
input3 = list(map(int, "1 10 2 9 3 8 4 7 5 6".split()))
```

```python
x = np.random.randint(-10, 10, 7).tolist()
print(x, solution.poorness(iter(x)))
```

```python
inputs = [input1, input2, input3]

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i, values in enumerate(inputs): 
    x = np.linspace(
        np.min(values), 
        np.max(values),
        num=10000
    )
    y = np.zeros(x.shape)
    for j, value in enumerate(x):
        y[j] = solution.weakness(value, values)

    ax[i].set_title(f"Weakness for Input {i + 1}")
    ax[i].plot(x, y)

fig.tight_layout()
plt.savefig("input.png", bbox_inches='tight', dpi=600)
```
